```python
import numpy as np
import pandas as pd

# 生成从-4到4间隔0.01的点
points = np.linspace(-8, 8, num=100)

# 创建DataFrame并保存为CS文件
df = pd.DataFrame(points)
df.to_csv('data/points.txt', header=False, index=False)

print("数据已保存到 points.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 读取数据文件（请替换为你的实际文件路径）
x_file = "/mnt/data2/home/syt/data/fhec/benchmark/moti/single/data/points.txt"
y1_file = "/mnt/data2/home/syt/data/fhec/benchmark/moti/single/data/res.txt"

try:
    # 使用numpy读取数据（适用于每行一个数值的文本文件）
    x = np.loadtxt(x_file)
    y1 = np.loadtxt(y1_file)
    
except Exception as e:
    print(f"读取文件失败: {e}")
    exit()

# 验证数据长度一致性
if not (len(x) == len(y1)):
    print("错误：数据长度不一致")
    exit()

# 创建画布
plt.figure(figsize=(10, 6), dpi=100)

# 绘制双曲线
plt.plot(x, y1, 
         color='#2B7FC3',    # 蓝色
         linewidth=2, 
         linestyle='-', 
         marker='o', 
         markersize=5,
         label='data1')

# 添加标注
plt.title("comp", fontsize=14, fontweight='bold')
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(frameon=True, shadow=True, loc='upper right')

# 自动调整坐标范围
plt.autoscale(enable=True, axis='both', tight=True)

# 显示图表
plt.tight_layout()
plt.show()

# 保存图片（可选）
# plt.savefig('/mnt/data2/home/syt/data/fhec/log/pic/comparison_plot.png', bbox_inches='tight')
```
