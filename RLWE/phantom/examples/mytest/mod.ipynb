```python
import random

def miller_rabin(n, k=5):
    if n == 2 or n == 3:
        return True
    if n < 2 or n % 2 == 0:
        return False
    
    # 将n-1写成2^r * d的形式
    r = 0
    d = n - 1
    while d % 2 == 0:
        r += 1
        d //= 2
    
    # 进行k次测试
    for _ in range(k):
        a = random.randrange(2, n-1)
        x = pow(a, d, n)
        if x == 1 or x == n-1:
            continue
        for _ in range(r-1):
            x = (x * x) % n
            if x == n-1:
                break
        else:
            return False
    return True

def generate_40bit_prime():
    while True:
        # 生成40位随机奇数
        n = random.getrandbits(40)
        # 确保是奇数
        n |= 1
        # 确保是40位
        n |= (1 << 39)
        if miller_rabin(n):
            return n

# 生成并打印一个40位素数
# prime = generate_40bit_prime()
# print(f"40位素数: {prime}")
# print(f"二进制长度: {len(bin(prime))-2}")
# print(f"二进制表示: {bin(prime)[2:]}")

# 生成5个不同的40位素数
primes = set()
while len(primes) < 5:
    primes.add(generate_40bit_prime())

for i, p in enumerate(primes, 1):
    print(f"素数 {i}: {p}")
```
