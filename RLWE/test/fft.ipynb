```python
## fft fft

import numpy as np

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.fft.fft(x)
print("FFT result:", y)

y1 = np.fft.fft(y)
print("FFT of FFT result:", y1)
##  y1[0,1,...,n-1] = n * x[0, n-1, n-2, ..., 1]

y2 = np.fft.ifft(x)
print("Inverse FFT result:", y2)

y3 = np.fft.ifft(y2)
print("Inverse FFT of Inverse FFT result:", y3)
##  y3[0,1,...,n-1] = 1/n * x[0, n-1, n-2, ..., 1]

y4 = np.fft.ifft(x)
y4 = np.fft.fft(y4)
print("FFT of Inverse FFT result:", y4)
```

```python
# ntt ntt

import galois as gs

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = gs.ntt(x, modulus=17)
print("NTT result:", y)

y1 = gs.ntt(y, modulus=17)
print("NTT of NTT result:", y1)
# y1[0,1,...,n-1] = n * x[0, n-1, n-2, ..., 1]

# N^{-1} = 15
y11 = np.array(y1) * 15 % 17
print("Inverse NTT result:", y11)

y2 = gs.intt(x, modulus=17)
print("Inverse NTT result:", y2)

y3 = gs.intt(y2, modulus=17)
print("Inverse NTT of Inverse NTT result:", y3)

y33 = np.array(y3) * 8 % 17
print("Inverse NTT of Inverse NTT result with scaling:", y33)
```
